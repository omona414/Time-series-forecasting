```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
ds = pd.read_csv('data/m2_s2_bank_0765f177a52e5543a9e0a8bfc045cf6a__site_40ab7c891c559b478e96f8bb5c682ca0.csv', encoding='utf8')
ds
```

```python
len_ds = len(ds)
len_ds
```

```python
max_cell = max(ds['CellNo'])
resist = [[None] for i in range(max_cell)]
resist[0]
```

```python
ds.loc[0,'CellNo']
```

```python
for i in range(len_ds):
    cell_no = ds.loc[i, 'CellNo']
    resist[(cell_no-1)].append(ds.loc[i, 'ResistValue'])
```

```python
for i in range(max_cell):
    resist[i] = list(filter(None.__ne__, resist[i]))
```

```python
resist = pd.DataFrame(resist)
resist
```

```python
resist = resist.dropna(axis=1)
resist
```

```python
resist = resist.transpose()
resist
```

```python
for i in range(max_cell):
    median = np.median(resist.loc[:,i])
    median_upper = 1.5*median
    median_lower = 0.5*median
    
    plt.figure(figsize=(9,6))
    plt.axhline(median, 0, 1, color='orange')
    plt.axhline(median_upper, 0, 1, color='lightgray', linestyle='--')
    plt.axhline(median_lower, 0, 1, color='lightgray', linestyle='--')
    plt.plot(resist.loc[:,i])
    plt.title(i)
    plt.show()
```
